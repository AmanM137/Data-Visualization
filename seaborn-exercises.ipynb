<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 1: <b>dataset -- titanic</b><br>
    Create the fare distribution through a histogram.</p>

```python
plt.figure(figsize=(6,4))
titanic = sns.load_dataset("titanic")
plt.hist(titanic[["fare"]], bins=30, edgecolor="grey", color="#ffabab")
ax=plt.gca()
plt.axis([0,602,-0.5,500])
plt.xlabel("fare")
plt.grid()
sp=["top","right","bottom","left"]
plt.tick_params(left = False,bottom=False)
ax.spines[sp].set_visible(False)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 2: <b>dataset -- titanic</b><br>
    Write a python program to create a boxplot for age and class group of all passengers.</p>

```python
plt.figure(figsize=(6,3))
sns.boxplot(data=titanic, x = titanic["class"], y=titanic["age"], palette="rainbow")
plt.axis([-0.5,2.5,-0.5,5])
plt.yticks(np.arange(0,90,10))
plt.grid(axis='y',alpha=0.6)
ax=plt.gca()
sp=["top","right","bottom","left"]
plt.tick_params(left = False,bottom=False)
ax.spines[sp].set_visible(False)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 3: <b>dataset -- titanic</b><br>
   Create a bar chart for the number of male and female passengers.</p>

```python
ax = titanic.groupby("sex")["sex"].count().sort_values(ascending=False).plot(kind="bar", width=0.9, color=[(0.2,0.4,0.7), (0.3,0.5,0.3)])
ax.grid(axis='y', which="both")
ax=plt.gca()
plt.axis([-0.6,1.6,-0.5,600.5])
sp=["top","bottom"]
plt.tick_params(left = False,bottom=False)
ax.spines[sp].set_visible(False)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 4: <b>dataset -- titanic</b><br>
    Set up a grid to plot survival probability against several variables like class, sex, who and alone.
    Then Draw a seaborn pointplot onto each Axes.</p>

```python
sns.set_style('darkgrid')
g=sns.PairGrid(titanic,x_vars=['class','sex','who','alone'],y_vars='survived', height=4.5, aspect=0.5)
g.map(sns.pointplot,color='#630330')
g.set(ylim=(0,1))
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 5: <b>dataset -- diamonds</b><br>
    Generate the distribution of prices with respect to color for each cut quality.</p>

```python
diamonds = sns.load_dataset("diamonds")
plt.figure(figsize=(5,4))
sns.barplot(data=diamonds, x = 'cut', y = 'price', hue='color', hue_order=['E', 'I', 'J', 'H', 'F', 'G', 'D'],
           order=['Ideal','Premium','Good','Very Good', 'Fair'])
plt.legend(ncol=4)
plt.xlabel("Cut",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 6: <b>dataset -- diamonds</b><br>
    Obtain the counts of diamonds of each cut quality through bar chart.</p>

```python
fig, ax = plt.subplots()
ax=sns.countplot(data=diamonds, x="cut", color="red", order=["Ideal", "Premium", "Good", "Very Good", "Fair"])
plt.grid(False)
ax.set_facecolor('white')
sp=["left","bottom"]
ax.spines[sp].set_color('black')
plt.tick_params(left = True,bottom=True)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 7: <b>dataset -- diamonds</b><br>
    Create a histogram of the frequency of diamonds in the dataset with their respective carat specifications on the x-axis.</p>

```python
ax=sns.distplot(x=diamonds["carat"], kde=False)
plt.grid(False)
ax.set_facecolor('white')
sp=["left","bottom","right","top"]
ax.spines[sp].set_color('black')
plt.axis([-0,5,0,12000])
plt.tick_params(left = True,bottom=True)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 8: <b>dataset -- tips</b><br>
    Draw a nested violinplot and split the violins for easier comparison between day and total_bill.</p>

```python
tips = sns.load_dataset("tips")
sns.set_style('whitegrid')
sns.violinplot(data=tips, x='day', y='total_bill',inner='quartile',palette=['#AECCE4','#FFFCBB'], hue='sex', split=True)
sns.despine(top=True, bottom=True, left=True, right=True)
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 9: <b>dataset -- titanic</b><br>
     Draw a nested barplot to show survival for class and sex</p>

```python
sns.barplot(data=titanic, x='class', y='survived', hue='sex')
plt.ylabel("survival probability")
plt.legend(title='sex', loc='center right', bbox_to_anchor=(1.2,0.5), frameon=False)
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 10: <b>dataset -- fmri</b><br>
    Create line plot to show the responses for different events and regions.</p>

```python
fmri = sns.load_dataset("fmri")
sns.set_style("darkgrid")
sns.lineplot(data=fmri, y='signal', x='timepoint', hue='region', style='event')
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 11: <b>dataset -- iris</b><br>
    Create a scatter plot between sepal length and sepal width categorised by species.</p>

```python
iris=sns.load_dataset("iris")
sns.set_style("ticks")
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
sns.despine(top=True, right=True)
plt.legend(loc="center right", frameon=False, bbox_to_anchor=(1.25,0.5))
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 12: <b>dataset -- iris</b><br>
    Visualize the kernel density estimate of petal length for each species.</p>

```python
for species in iris['species'].unique():
    sns.kdeplot(data=iris[iris['species']==species], x='petal_length', label=species)
sns.despine(top=True, right=True)
plt.ylabel("")
plt.legend(loc="center right", frameon=False, bbox_to_anchor=(1.25,0.5))
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 13: <b>dataset -- iris</b><br>
    Study the relationship between sepal length and sepal width for the iris dataset by creating a joint plot.</p>

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', marginal_kws={'color':"#6fa8dc"})
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 14: <b>dataset -- iris</b><br>
    Create a paiplot for the dataset to show relation between each pair of features based on different species.</p>

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

<p style="text-align:left; font-family:newtimeroman;padding-left:10px;padding-right:10px;font-size:110%;margin:0px">Ques 15: <b>dataset -- exercise</b><br>
    Draw a pointplot to show pulse as a function of three categorical factors rest, walking and running.</p>

```python
exercise=sns.load_dataset("exercise")
#plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
g = sns.FacetGrid(exercise, col='diet')
g.map(sns.pointplot, 'time', 'pulse', 'kind',hue_order=['rest', 'walking', 'running'])
plt.legend(frameon=False, loc="center right", bbox_to_anchor=(1.4,0.5))
plt.show()
```
